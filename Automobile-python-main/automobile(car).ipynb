## Visualizing the distribution of a dataset



```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
auto = pd.read_csv('Automobile.csv.csv')
```

```python
auto.head()
```

## Plotting univariate distributions



```python
sns.distplot(auto['price'])
```

## Histograms

Histograms are likely familiar, and a hist function already exists in matplotlib. 


To illustrate this, let’s remove the density curve and add a rug plot

```python
sns.histplot(auto['city_mpg'], kde=False);
```

## Plotting bivariate distributions

It can also be useful to visualize a bivariate distribution of two variables. The easiest way to do this in seaborn is to just use the jointplot() function


```python
sns.jointplot(auto['price'], auto['horsepower'],kind='reg',color='green')
```

### Hex Bin Plots

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="hex",color='pink')
```

### Kernel Density Estimation

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="hist")
```

## Visualizing pairwise relationships in a dataset

To plot multiple pairwise bivariate distributions in a dataset
and we use the pairplot() function.
This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame.

```python
sns.pairplot(auto[['price', 'engine_size', 'horsepower']])
```

## stripplot fuel and horsepower 



```python
sns.stripplot(auto['fuel_type'], auto['horsepower'])
```

A different approach would be to use the function swarmplot(), which positions each scatterplot point on the categorical axis with an algorithm that avoids overlapping points:

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower'])
```

## Boxplots


```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['body_style'])
```

## Bar plots



```python
sns.barplot(auto['body_style'], auto['horsepower'], hue=auto['engine_location'])
```

```python
sns.countplot(auto['body_style'])
```

## Point plots



```python
sns.pointplot(auto['fuel_system'], auto['horsepower'], hue=auto['number_of_doors'])
```

## Drawing multi-panel categorical plots

```python
sns.factorplot(x="fuel_type", 
               y="horsepower", 
               hue="number_of_doors", 
               col="engine_location", 
               data=auto, 
               kind="swarm")
```

## Function to draw linear regression models

lmplot() is one of the most widely used function to quickly plot the Linear Relationship b/w 2 variables

```python
sns.lmplot(x="horsepower", y="price", data=auto)
```

```python
sns.lmplot(x="horsepower", y="price",hue="fuel_type", data=auto)
```

```python
sns.lmplot(x="horsepower", y="price",hue="number_of_doors", data=auto)
```
